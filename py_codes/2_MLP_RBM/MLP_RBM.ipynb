```python
import keras
from sklearn.neural_network import BernoulliRBM
import numpy as np
from sklearn import linear_model, datasets, metrics
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.neural_network import MLPClassifier
import pickle
from sklearn.base import clone
import pandas as pd
from keras.datasets import mnist

digits = datasets.load_digits()
X = np.asarray(digits.data, 'float32')
Y = digits.target
X = (X - np.min(X, 0)) / (np.max(X, 0) + 0.0001)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
                                              test_size=0.2,
                                                    random_state=0)
clf = MLPClassifier(activation='tanh',
              hidden_layer_sizes=(512, 512),
              learning_rate='adaptive', learning_rate_init=0.01,
              max_iter=30,
              solver='sgd',
              verbose=True)
rbm1 = BernoulliRBM(n_components=50, learning_rate=0.06, n_iter=20, verbose=True, random_state=0)
rbm2 = BernoulliRBM(n_components=32, learning_rate=0.06, n_iter=40, verbose=True, random_state=0)
DBN3 = Pipeline(steps=[('rbm1', rbm1), ('rbm2', rbm2), ('clf', clf)])
hist = DBN3.fit(X_train, Y_train)
print("Classification using RBM features:\n%s\n" % (
    metrics.classification_report(
        Y_test,
        DBN3.predict(X_test))))
y_pred = DBN3.predict(X_test)
# print(sklearn.metrics.accuracy_score(Y_test, y_pred, normalize=True, sample_weight=None))
# confusion_mat = np.zeros((10,10) ,  dtype=np.int32)
# for j in range(0, 360):
#     confusion_mat[y_pred[j], Y_test[j]] = confusion_mat[y_pred[j], Y_test[j]] +1;   
# print(confusion_mat)
```
