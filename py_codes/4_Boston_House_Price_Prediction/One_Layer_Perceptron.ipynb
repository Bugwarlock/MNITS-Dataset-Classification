```python
from keras.models import Sequential
from keras.layers.core import Dense
from keras.optimizers import SGD, Adam
from keras.initializers import RandomNormal
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = np.array(pd.read_excel(r'Boston.xlsx'))
x_test = data[100:,:-1] 
y_test = data[100:,-1]
x_train = data[:100,:-1]
y_train = data[:100,-1]
for i in range(len(y_train)):
    if(np.isnan(y_train[i])):
        y_train[i] = 20+(np.random.rand()*10)
for i in range(len(x_train)):
    for j in range(13):
        if(np.isnan(x_train[i,j])):
            x_train[i,j] = 20+(np.random.rand()*10)
for i in range(len(y_test)):
    if(np.isnan(y_test[i])):
        y_test[i] = 20+(np.random.rand()*10)
for i in range(len(x_test)):
    for j in range(13):
        if(np.isnan(x_test[i,j])):
            x_test[i,j] = 20+(np.random.rand()*10)        

model = Sequential()
model.add(Dense(20, input_shape=(13, ), activation='softmax'))
# model.add(Dense(15, activation='softmax'))
model.add(Dense(1,activation='relu'))
sgd = SGD(lr=0.01)
OPT = Adam(lr=0.1)
model.compile(loss='mean_squared_error', optimizer='sgd',
             metrics = ['accuracy'])
n_epochs = 30
hist = model.fit(x_train, y_train,
                 epochs=n_epochs,
                 batch_size=1,
                 verbose=1, validation_split=0.2)
model.summary()
te_score = model.evaluate(x_test, y_test, verbose = 0)
print('Test Loss:', te_score[0])
print('Test accuracy:', te_score[1])
train_loss = hist.history['loss']
val_loss = hist.history['val_loss']
xc=range(n_epochs)
plt.figure(1, figsize=(7, 5))
plt.plot(xc, train_loss)
plt.plot(xc, val_loss)
plt.xlabel('num of Epochs')
plt.ylabel('loss')
plt.title('train_loss vs val_loss')
plt.grid(True)
plt.legend(['train', 'val'])
#print plt.style.available # use bmh, classic,ggplot for big pictures
plt.style.use(['classic'])
plt.show()

```
